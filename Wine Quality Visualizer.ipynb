```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df= pd.read_csv("C:\\Users\\schne\\Desktop\\Project2\\winequality-red.csv")
```

## What Best Determines Wine Quality?
The dataset "winewuality-red.csv" was taken off Kaggle, and I was looking to see which variables best determines quality in the dataset. 
Starting off is a bar plot showing the number of wines at each quality rating. 
The Quality being based on sensory data on a score of 0-10, however no wine exceeds a score of 8 and no wine scoring less than a 3. 

```python
ax = df["quality"].value_counts()
x = [3,4,5,6,7,8]
y = ax.sort_index(axis=0, ascending=True)
plt.bar(x,y)
plt.xlabel("Quality Rating")
plt.ylabel("Number of Instances")
plt.title("Number of instances of each Quality Rating")
plt.show
```

Here on this graph we see most the wines are rated at 5 and 6 with the rest leaning towards the higher end of the scale.

## Alcohol Content vs Sugar Content

This graph takes the Alchohol content and plots it against the sugar content with the quality showing up in the color of the dots. This graph was to see whether or not the combination of a sweet wine and the alcohol content affected the quality score that the wine was given.

```python
x = df.loc[:,"residual sugar"]
y = df.loc[:,"alcohol"]
colors = df.loc[:,"quality"]
plt.scatter(x,y,c=colors)
cbar = plt.colorbar()
cbar.set_label("Quality of the Wine")
plt.xlabel("Residual Sugar Content")
plt.ylabel("Alcohol Content")
plt.title("Alcohol Content vs Residual Sugar Content")
plt.show()
```

As shown some of the best wines have a higher alcohol content, but are lower on the residual sugar content. This would make sense as the people rating the wines probably don't like sugar covering up the flavor of the wine. With some of the lowest rated wines coming in with a lower alcohol content.

##  pH vs Desnity

This graph we take a look at pH vs Density, seeing if the level ef acidity(pH) and the density has an effect on the quality. pH the lower the number the more acidic it is, with most wines being between 3-4. The density of the dataset is how close it resemembles the density of water.

```python
x = df.loc[:,"density"]
y = df.loc[:,"pH"]
colors = df.loc[:,"quality"]
plt.scatter(x,y,c=colors)
cbar = plt.colorbar()
cbar.set_label("Quality of the Wine")
plt.xlabel("Density")
plt.ylabel("pH level")
plt.title("pH level vs density of the wine")
plt.show()
```

As shown in the graph the wine that is less dense or feels "lighter" typically has a better score than wine that has a similar or heavier density to water. Where pH level there doesn't seem to be too much correlation but a very slight one showing that a higher pH level indicates a better wine.

## Fixed Acidity vs Citric Acid

This graph shows the fixed acidity which is the acids in the wine that do not evaporate readily, vs the citric acid which can add "freshness and flavor to wines" 

```python
x = df.loc[:,"citric acid"]
y = df.loc[:,"fixed acidity"]
colors = df.loc[:,"quality"]
plt.scatter(x,y,c=colors)
cbar = plt.colorbar()
cbar.set_label("Quality of the Wine")
plt.xlabel("Citric Acid")
plt.ylabel("Fixed Acidity")
plt.title("Fixed Acidity vs Citric Acid")
plt.show()
```

As shown in the graph you can see that the higher the citric acid normally leads to a better quality of wine. With the fixed Acidity not really have a census on the quality of the wine.
